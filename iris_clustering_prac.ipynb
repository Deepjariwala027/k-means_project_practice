```python
#importing all the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
sns.set()
```

```python
#importing the dataset
data_set = pd.read_csv('iris-dataset.csv')
data_set
```

```python
#scatterplot to visualize the data for sepal
plt.figure(figsize=(12,12))
plt.scatter(data_set['sepal_length'],data_set['sepal_width'])
plt.show
```

```python
#scatterplot to visualize the data for petal
plt.figure(figsize=(12,12))
plt.scatter(data_set['petal_length'],data_set['petal_width'])
plt.show
```

```python
# iloc is a method used to 'slice' data 
# 'slice' is not technically correct as there are methods 'slice' which are a bit different
# The term used by pandas is 'selection by position'
# The first argument of identifies the rows we want to keep
# The second - the columns
# When choosing the columns, e.g. a:b, we will keep columns a,a+1,a+2,...,b-1 ; so column b is excluded
x1 = data_set.iloc[:,0:2]
# for this particular case, we are choosing columns 1 and 2
# Note column indices in Python start from 0
x1
```

```python
# iloc is a method used to 'slice' data 
# 'slice' is not technically correct as there are methods 'slice' which are a bit different
# The term used by pandas is 'selection by position'
# The first argument of identifies the rows we want to keep
# The second - the columns
# When choosing the columns, e.g. a:b, we will keep columns a,a+1,a+2,...,b-1 ; so column b is excluded
x2 = data_set.iloc[:,2:4]
# for this particular case, we are choosing columns 1 and 2
# Note column indices in Python start from 0
x2
```

```python
#applying kmeans algorithm and identifying clusters
kmeans_1 = KMeans(2)
kmeans_1.fit(x1)

identified_clusters_1 = kmeans_1.fit_predict(x1)
print(identified_clusters_1)

# Create a copy of the data
data_with_clusters = data_set.copy()
# Create a new Series, containing the identified cluster for each observation
data_with_clusters['Cluster1'] = identified_clusters_1
# Check the result
data_with_clusters
```

```python
#applying kmeans algorithm and identifying clusters
kmeans_2 = KMeans(3)
kmeans_2.fit(x2)

identified_clusters_2 = kmeans_2.fit_predict(x2)
print(identified_clusters_2)

# Create a copy of the data
data_with_clusters_2 = data_with_clusters.copy()
# Create a new Series, containing the identified cluster for each observation
data_with_clusters_2['Cluster2'] = identified_clusters_2
# Check the result
data_with_clusters_2
```

```python
plt.figure(figsize=(12,12))
plt.scatter(data_with_clusters['sepal_length'],data_with_clusters['sepal_width'],c=data_with_clusters['Cluster1'],cmap='rainbow')
plt.xlabel('Lenght of sepal')
plt.ylabel('Width of sepal')
plt.show()
```

```python
plt.figure(figsize=(12,12))
plt.scatter(data_with_clusters['petal_length'],data_with_clusters_2['petal_width'],c=data_with_clusters_2['Cluster2'],cmap='gist_rainbow')
plt.xlabel('Lenght of petal')
plt.ylabel('Width of petal')
plt.show()
```

```python
#wcss for the current value of k
kmeans_1.inertia_
```

```python
#wcss for the current value of k
kmeans_2.inertia_
```

```python
# THE ELBOW METHOD for sepal
wcss_1=[]

#creating all possible cluster solutions
for i in range(1,7):
    #cluster solution of ith cluster
    kmeans_1 = KMeans(i)
    #fitting tha data
    kmeans_1.fit(x1)
    #wcss for each iteration and appending the values
    wcss_iter = kmeans_1.inertia_
    wcss_1.append(wcss_iter)

wcss_1
```

```python
# THE ELBOW METHOD for petal
wcss_2=[]

#creating all possible cluster solutions
for i in range(1,7):
    #cluster solution of ith cluster
    kmeans_2 = KMeans(i)
    #fitting tha data
    kmeans_2.fit(x2)
    #wcss for each iteration and appending the values
    wcss_iter = kmeans_2.inertia_
    wcss_2.append(wcss_iter)

wcss_2
```

```python
plt.figure(figsize=(12,12))
# Create a variable containing the numbers from 1 to 6, so we can use it as X axis of the future plot
number_clusters = range(1,7)
# Plot the number of clusters vs WCSS
plt.plot(number_clusters,wcss_1)
# Name your graph
plt.title('The Elbow Method')
# Name the x-axis
plt.xlabel('Number of clusters')
# Name the y-axis
plt.ylabel('Within-cluster Sum of Squares')

# Create a variable containing the numbers from 1 to 6, so we can use it as X axis of the future plot
number_clusters = range(1,7)
# Plot the number of clusters vs WCSS
plt.plot(number_clusters,wcss_2)
# Name your graph
plt.title('The Elbow Method')
# Name the x-axis
plt.xlabel('Number of clusters')
# Name the y-axis
plt.ylabel('Within-cluster Sum of Squares')
```

```python
# same code using standardization techniques
from sklearn import preprocessing

#scaling the values
x1_scaled = preprocessing.scale(x1)
x1_scaled
```

```python
#kmeans on scaled dataset
kmeans_1_scaled = KMeans(2)
kmeans_1_scaled.fit(x1_scaled)

# create a copy of data, so we can see the clusters next to the original data
clusters_scaled = data_set.copy()
# predict the cluster for each observation
clusters_scaled['cluster_pred']=kmeans_1_scaled.fit_predict(x1_scaled)

# create a scatter plot based on two corresponding features (sepal_length and sepal_width; OR petal_length and petal_width)
plt.scatter(clusters_scaled['sepal_length'], clusters_scaled['sepal_width'], c= clusters_scaled ['cluster_pred'], cmap = 'rainbow')
```
